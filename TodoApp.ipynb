```python
toDoList = [] 
def add_element() :
    add = input("Enter the data")
    toDoList.append(add)
def del_element() :
    pos = int(input("Enter the position you want to delete\n"))
    del toDoList[pos]
def Clear() :
    toDoList.clear()
while True :
    selection = input("Enter your Selection ID\n1.Add Element\n2.Delete Element\n3.Delete all\n4.exit\n")
    if selection == "1" :
        add = input("Enter the data\n")
        toDoList.append(add)
    elif selection == "2" :
        del_element()
    elif selection == "3" :
        Clear()
    elif selection == "4" :
        break
    else :
        print("Enter correct ID")
    print(toDoList)
    print("\t\t\tXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX\t\t\t")
```

```python
toDoData = {} 
menu = {
    1 : "Add Element",
    2 : "View Elements",
    3 : "Delete Element",
    4 : "Delete All Elements",
    5 : "Terminate App",
}

def isInteger(check) :
    try :
        int(check)
        return True
    except ValueError :
        return False

def mainMenu(menu) :
    print("\nXXXXXXXXXXXXXXXXXXXX\n        Menu        \nXXXXXXXXXXXXXXXXXXXX\n")
    for key,value in menu.items() :
        print("{}. {}".format(key,value))
    userCommand = input("Enter Command \n")
    return userCommand

def add_items(toDoData) :
    print("\nXXXXXXXXXXXXXXXXXXXX\n     Add Item       \nXXXXXXXXXXXXXXXXXXXX\n")
    toDoIndex = len(toDoData) + 1
    itemData = input("Item{}\n".format(toDoIndex))
    toDoData.update({toDoIndex:itemData})  
    
def view_items(toDoData) :
    print("\nXXXXXXXXXXXXXXXXXXXX\n     View Items        \nXXXXXXXXXXXXXXXXXXXX\n")
    if len(toDoData) > 0 :
        for key, value in toDoData.items() :
            print("{}.{}".format(key,value))
    else :
        print("No Task Found\n")

def delete_items(toDoData) :
    print("\nXXXXXXXXXXXXXXXXXXXX\n   Delete Items        \nXXXXXXXXXXXXXXXXXXXX\n")
    if len(toDoData) > 0 :
        for key, value in toDoData.items() :
            print("{}.{}".format(key,value))
        del_index = int(input("Enter Deleting Index eg: 1"))
        if del_index in toDoData.keys() :
            del toDoData[del_index]
        else :
            print("Invalid Index\n")
    else :
        print("No Task Found\n")  
        
def delete_all_items(toDoData) :
    print("\nXXXXXXXXXXXXXXXXXXXX\n Delete All Items        \nXXXXXXXXXXXXXXXXXXXX\n")
    if len(toDoData) > 0 :
        toDoData.clear()
        print("Mission Successful All Items deleted\n")
            
while True :
    userCommand = mainMenu(menu)
    if isInteger(userCommand) == True :
        userCommand = int(userCommand)
        if userCommand > 0 and userCommand <= 5 :
            
            if userCommand == 1 :
                add_items(toDoData)
            
            elif userCommand == 2 :
                view_items(toDoData)
                
            elif userCommand == 3 :
                delete_items(toDoData)
                
            elif userCommand == 4 :
                delete_all_items(toDoData)
                
            elif userCommand == 5 :
                print("App Terminated")
                break
            else :
                print("Not Matched\n")
            
        else :
            print("Invalid Input, Out of Range must in be 0 to 5")
    else :
        print("Invalid Input, Try Again\n")
```
